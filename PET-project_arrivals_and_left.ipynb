```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Зчитано таблицю кількості прибулих в Україну в 2021 році для подальших групувань
```

```python
arrivals = pd.read_excel('C:\\Users\\User\\Desktop\\PET_project\\Кількість_прибулих_2021.xlsx')
```

```python
print(arrivals)
```

```python
# Згруповано дані за віком з просумуванням значень в стовпці "2021", індекс скинуто для отримання нового DataFrame з рещультатами
```

```python
grouped_arrivals = arrivals.groupby('Вік')['2021'].sum().reset_index()
```

```python
# Дані відсортовано за спаданням по значенню в стовпці "2021"
```

```python
grouped_arrivals_sort = grouped_arrivals.sort_values (by = '2021', ascending = False)
```

```python
print(grouped_arrivals_sort)
```

```python
# Збережено таблицю в форматі xlsx
```

```python
grouped_arrivals_sort.to_excel('Кількість прибулих в 2021.xlsx', index = False)
```

```python
# Обчислено загальну кількість прибулих за статтю у 2021 році
```

```python
total_arrivals = arrivals.groupby('Стать')['2021'].sum()
```

```python
print(total_arrivals)
```

```python
# Розраховано загальний коефіцієнт міграції для прибулих
```

```python
migration_ratio_arrivals = round(total_arrivals ['Чоловіки']/total_arrivals ['Жінки'], 2)
```

```python
print(f"Коефіцієнт міграції для прибулих:{migration_ratio_arrivals}")
```

```python
# Створено словник для зберігання даних
```

```python
age_groups = {}
```

```python
# Заповнено словник
```

```python
for gender, age_group, count in data:
    if age_group not in age_groups:
        age_groups[age_group] = {"Чоловіки": 0, "Жінки": 0}
    age_groups[age_group][gender] += count
```

```python
# Обчислено і виведено коефіцієнти
```

```python
for age_group, counts in age_groups.items():
    if counts["Жінки"] > 0:
        ratio = counts["Чоловіки"] / counts["Жінки"]
        print(f"Вікова категорія {age_group}: коефіцієнт = {ratio:.2f}")
    else:
        print(f"Вікова категорія {age_group}: недостатньо даних про жінок")
```
